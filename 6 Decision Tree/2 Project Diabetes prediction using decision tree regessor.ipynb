## Project Diabetes prediction using decision tree regessor:-

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data,
                 columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df_diabetes.head()
```

### Dependant and Independant features:

```python
X = df_diabetes
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.head()
```

```python
y_train
```

### Check Correlation:

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(X_train.corr(),annot=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

### Hyperparameter tuning:

```python
param = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth' :[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt', 'log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor,param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(f'R2 score: {r2_score(y_test,y_pred)}')
print(f'MAE: {mean_absolute_error(y_test,y_pred)}')
print(f'MSE: {mean_squared_error(y_test,y_pred)}')
```

### To check decision tree with best parameters after hyperparametere tunning:

```python
selectmodel = DecisionTreeRegressor(criterion='friedman_mse',max_depth= 5,max_features= 'sqrt',splitter= 'random')
```

```python
selectmodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,8))
tree.plot_tree(selectmodel,filled=True)
plt.show()
```
