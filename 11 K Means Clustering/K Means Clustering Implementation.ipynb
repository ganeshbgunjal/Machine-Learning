```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
X,y = make_blobs(n_samples=1000,centers=3,n_features=2)
```

### Standardization:-- Feature scaling technique

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

#### Here one imp thing is we are not going to use y_train and y_test. because there is no need of this. we are taking only indepedant feaures. not that ouput or target feature.

### Elbow method to select K value:

```python
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss    #as k values increases, wcss decreases
```

#### Now how to determine what should be the exact K values

### Plot Elbow curve:

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
kmeans.fit_predict(X_train_scaled)
```

```python
y_pred = kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
plt.show()
```

### Silhoutee Scoring:

```python
from sklearn.metrics import silhouette_score
```

```python
silhoutee_coefficient = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train_scaled)
    score = silhouette_score(X_train_scaled,kmeans.labels_)
    silhoutee_coefficient.append(score)
    
```

```python
silhoutee_coefficient
```

```python
plt.plot(range(2,11),silhoutee_coefficient)
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhoutee coefficient')
plt.show()
```

#### wherever the value is high for silhoutee score, use that valule to decide K value
