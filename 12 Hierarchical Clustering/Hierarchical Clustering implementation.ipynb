```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

### Standardisation:-

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

### Apply PCA:--

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=iris.target)
plt.show()
```

### Apply Agglomerative Clustering:
### 1. Constract a Dendogram:--

```python
import scipy.cluster.hierarchy as sc

# Plot the dendogram:
plt.figure(figsize=(20,7))
plt.title('Dendrogram')

sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.xlabel('Sample index')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
aggl_cluster = AgglomerativeClustering(n_clusters=2, metric= 'euclidean',linkage='ward')
aggl_cluster.fit(pca_scaled)
```

```python
aggl_cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c= aggl_cluster.labels_)
plt.show()
```
