we should remove highly correlated feature.  
Highly correlated features are two or more input variables (features) in a dataset that show a strong statistical relationship with each other.

Types of Correlation

Positive correlation → As one feature increases, the other also increases.
Example: Height and weight usually have a positive correlation.

Negative correlation → As one feature increases, the other decreases.
Example: Speed of a car and travel time for the same distance.

How to Measure Correlation

Correlation coefficient (Pearson’s r): Ranges from -1 to +1

+1 → Perfect positive correlation

-1 → Perfect negative correlation

0 → No correlation
