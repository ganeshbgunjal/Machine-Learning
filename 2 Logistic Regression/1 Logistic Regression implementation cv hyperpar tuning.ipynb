## Logistic Regression practical implementation:

```python
import pandas as pd
import numpy as np
import seaborn as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

make_classification() helps you generate a synthetic classification dataset — very useful for practicing machine learning algorithms without needing a real-world dataset.

### Create the dataset using above make classification library:

```python
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

### Model training

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(f'Accuracy score: {score}')
cm = confusion_matrix(y_test,y_pred)
print(f'Confusion matrix: {cm}')
print(classification_report(y_test,y_pred))
```

### Hyperparameter tuning and cross validation:

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

### Hyper parameter tuning technique: 
### 1. GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy',cv=cv, n_jobs=-1)

# n_jobs = -1 means it will all the processors.
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
print(f'Accuracy score: {score}')
cm = confusion_matrix(y_test,y_pred)
print(f'Confusion matrix: {cm}')
print(classification_report(y_test,y_pred))
```

### 2 Randomized search CV:

RandomizedSearchCV is used to find the best hyperparameters for a machine learning model — just like GridSearchCV, but faster because it randomly samples a few combinations instead of testing all possible ones.

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
y_pred = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
print(f'Accuracy score: {score}')
cm = confusion_matrix(y_test,y_pred)
print(f'Confusion matrix: {cm}')
print(classification_report(y_test,y_pred))
```

#### this technique useful when we have large dataset. as it take samples and not the whole data.
