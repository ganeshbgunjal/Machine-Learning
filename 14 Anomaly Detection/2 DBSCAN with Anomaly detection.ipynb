## DBSCAN Clutering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.10)
```

```python
dbscan.fit_predict(X)
```

#### -1 means outliers in above values.

#### The method fit_predict() is a combination of two steps  fit() and predict()  done in a single call, commonly used in unsupervised learning algorithms such as K-Means, DBSCAN, or Gaussian Mixture Models.

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
plt.show()
```
