## Local Outlier factor Anomaly Detection:

### ðŸ§  What is Local Outlier Factor (LOF)?

LOF detects anomalies by comparing the local density of each point to that of its neighbors.

Points that have much lower density than their neighbors are considered outliers.

Itâ€™s an unsupervised anomaly detection method.

```python
from sklearn.neighbors import LocalOutlierFactor
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
# Step 2: Create circular data
X,y = make_circles(n_samples=750,factor=0.3,noise=0.1,random_state=42)
```

```python
X
```

```python
# Step 3: Fit the Local Outlier Factor model
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
```

```python
y_pred = lof.fit_predict(X)
```

```python
# Step 4: Identify inliers and outliers
# LOF returns 1 for inliers, -1 for outliers

inliers = X[y_pred == 1]
outliers = X[y_pred == -1]
```

```python
# Step 5: Visualization
plt.figure(figsize=(8,6))
plt.scatter(inliers[:, 0], inliers[:, 1], c='blue', edgecolor='k', s=30, label='Inliers')
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', edgecolor='k', s=50, label='Outliers')
plt.title("Local Outlier Factor (LOF) Anomaly Detection", fontsize=14)
plt.legend()
plt.show()
```
